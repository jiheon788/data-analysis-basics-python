# Seaborn

## 0. PDF 확률 밀도 함수
* 확률 밀도 함수는 데이터셋의 분포를 나타낸다.
* 특정 구간 확률 = 구간의 면적
* 그래프 아래 모든 면적의 합은 1
* 특정 값의 확률이 아닌, 특정 구간의 확률

```python
import pandas as pd
import seaborn as sns
```

```python
!pip show seaborn
```

```python
bodyDf = pd.read_csv('data/body.csv',index_col=0)
bodyDf
```

## 1. KDE Plot
* 데이터의 분포를 확인하는 커널 밀도 추정(KDE: Kernel Density Estimator) 플롯
* bin width 줄이면 정교해짐
* bin width 키우면 스무스해지지만 정확도 낮아짐

```python
bodyDf['Height'].value_counts().sort_index().plot() #인덱스 순으로 정렬
```

```python
sns.kdeplot(bodyDf['Height'], bw=0.5) 
```

```python
bodyDf.plot(kind='hist',y='Height',bins=15)
```

```python
sns.distplot(bodyDf['Height'],bins=15)
```

```python
bodyDf.plot(kind='box',y='Height')
```

```python
sns.violinplot(y=bodyDf['Height'])
```

```python
bodyDf.plot(kind='scatter',x='Height',y='Weight')
```

```python
sns.kdeplot(bodyDf['Height'], bodyDf['Weight'])
```

## 2. LM PLot

```python
bodyDf.plot(kind='scatter', x='Height', y='Weight')
```

```python
sns.lmplot(data=bodyDf, x='Height', y='Weight')
#회귀선
```

## 3. Cat Plot

```python
laptopsDf = pd.read_csv('data/laptops.csv')
laptopsDf
```

```python
laptopsDf['os'].unique()
```

```python
sns.catplot(data=laptopsDf, x='os', y='price', kind='box')
```

```python
sns.catplot(data=laptopsDf, x='os', y='price', kind='violin')
```

```python
sns.catplot(data=laptopsDf, x='os', y='price', kind='strip')
```

```python
laptopsDf['processor_brand'].unique()
```

```python
sns.catplot(data=laptopsDf, x='os', y='price', kind='strip',hue='processor_brand')
```

```python
sns.catplot(data=laptopsDf, x='os', y='price', kind='swarm',hue='processor_brand')
```
