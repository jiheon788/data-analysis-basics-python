## 이상값 다루기

```python
df = pd.read_csv('data/beer.csv',index_col=0)
df
```

```python
df.plot(kind='box', y='abv')
```

```python
df['abv'].describe()
```

```python
q1=df['abv'].quantile(0.25)
```

```python
q3=df['abv'].quantile(0.75)
```

```python
iqr=q3-q1
iqr
```

```python
(df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
```

```python
condition=(df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
df.loc[2250,'abv']=0.055
df.loc[2250]
```

```python
condition=(df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index,inplace=True)
```

```python
condition=(df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
df.plot(kind='box', y='abv')
```
